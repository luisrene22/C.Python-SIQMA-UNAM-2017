```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib auto
%matplotlib inline
```

```python
def F(x,y):
    return np.array([-y,x])
```

```python
x = np.linspace(-2,2,11)
y = np.linspace(-2,2,11)
xmesh,ymesh = np.meshgrid(x,y)
```

```python
umesh,vmesh = F(xmesh,ymesh)
```

```python
fig = plt.figure()
ax = plt.subplot()
ax.quiver(xmesh,ymesh,umesh,vmesh)
plt.show()
```

```python
def σ(t):
    return np.array([t*np.sin(t),t*np.cos(t),t])
```

```python
t = np.linspace(0,10,51)
```

```python
x,y,z = σ(t)
```

```python
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x,y,z)
plt.show()
```

```python
def F(x,y,z):
    return np.array([-y,x,np.sin(z)])
```

```python
x = np.linspace(-5,5,11)
y = np.linspace(-5,5,11)
z = np.linspace(-2,2,3)
xmesh,ymesh,zmesh = np.meshgrid(x,y,z)
```

```python
umesh,vmesh,wmesh = F(xmesh,ymesh,zmesh)
```

```python
from mplt_toolkits.mplot3d import axes3d
fig = plt.figure()
ax=fig.gca(projection='3d')
ax.quiver(xmesh,ymesh,zmesh,umesh,vmesh,wmesh,length=0)
plt.show()
```
