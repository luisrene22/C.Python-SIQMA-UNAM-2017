```python
#campo de direcciones
def F(u,t,c=0.25,m=1.,k=1.):
    return np.array([u[1],-c*u[1]/m-k*u[0]/m])

#Condicion inicial
t = 0.
x = 2.
v=0.
u=np.array([x,v])

# Valores de ciclo
tfin=100.
dt=0.1

#solucion
tsol=[t]
xsol=[x]
vsol=[v]

# sol tiempo de Euler
while t<=tfin:
    u =F(u,t)*dt+u
    t=t+dt
    print(t)
    tsol.append(t)
    xsol.append(u[0])
    vsol.append(u[1])

fig1=plt.figure()
ax=plt.subplot(111)
ax.plot(tsol,xsol)
plt.show()
```

```python
#sistema
L = 10

#condiciones de frontera
T0 = 25.
TL = 1000.

#discretizar
N = 15

x = np.linspace(0,L,N)
dx = L/(N-1)

#Llenado de sistema de ecuaciones
A = np.zeros((N,N))
b = np.zeros(N)

for i in range(N):
    if i == 0:
        A[i][i]=1.
        b[i]=T0
    elif i == N-1:
        A[i][i]=1.
        b[i]=TL
    else:
        A[i][i] = -2./dx**2
        A[i][i+1] = 1./dx**2
        A[i][i-1] = 1./dx**2

#Sol sist ec

Ainv = np.linalg.inv(A)
T = np.dot(Ainv,b)

#graficas

plt.plot(x,T,'ro')
plt.grid(True)

plt.plot(x,T,'b')
plt.grid(alpha=0.3)

plt.show()
```
