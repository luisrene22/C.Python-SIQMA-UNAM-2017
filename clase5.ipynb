````python
class Carro():
    
    def __init__(self,color,cilindros):
        self.color = color
        self.cilindros = cilindros
    
    def encender(self):
        print('Carro Encendido')

class Ford(Carro):
    def encender(self):
        print('For Encendido')

class Toyota(Carro):
    def __init__(self,color,cilindros,altura):
        super(Toyota,self).__init__(color,cilindros) # ---> Super?
        self.altura = altura
        
    def encender(self):
        super(Toyota,self).encender()
        print('Toyota encendido')
    
T1 = Toyota('negro',8,3)
T1.encender()
````

# Generadores

````python
def recta(xinc,xfin):
    x = xinc
    while x <= xfin:
        y = 2*x+1
        yield (x,y) #yield es un return para un iterador
        x = x+1
for x,y in recta(0,1000):
    print(x,y)
    
````
